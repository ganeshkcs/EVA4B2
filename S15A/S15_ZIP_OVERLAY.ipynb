```
black_image = np.zeros((224,224))
bg_fg_counter = 1
bg_fg_mask_counter = 1
initial_bg_number = 1
print(datetime.datetime.now())
for bg_number in range(1,101):

  print("Bg Loop",datetime.datetime.now())
  bg = Image.open(f'/content/drive/My Drive/Utils/S15A/Images/background/{str(bg_number)}.jpg')

  zip_fg_bg = ZipFile(f'/content/drive/My Drive/Utils/S15A/Images/fg_bg{str(bg_number)}.zip', mode='a', compression=zipfile.ZIP_STORED)
  zip_fg_bg_mask = ZipFile(f'/content/drive/My Drive/Utils/S15A/Images/fg_bg_masks{str(bg_number)}.zip', mode='a', compression=zipfile.ZIP_STORED)

  # if initial_bg_number != bg_number:
  #   zip_fg_bg.close()
  #   zip_fg_bg_mask.close()
  #   initial_bg_number = bg_number
  fg_directory_name = "fg_bg" + str(bg_number)
  fg_mask_directory_name = "fg_bg_masks" + str(bg_number)

  # os.mkdir(fg_directory_name)
  # os.mkdir(fg_mask_directory_name)

  print("directory name", fg_directory_name )
  print("bg ", bg_number)

  for fg_number in range(1,101):
    fg = Image.open(f'/content/drive/My Drive/Utils/S15A/Images/foreground/{str(fg_number)}.png')
    fg_mask = Image.open(f'/content/drive/My Drive/Utils/S15A/Images/fg_masks/{str(fg_number)}.jpg').convert('1')

    for k in range(1,21):

      x_cord = random.randint(1, 101)
      y_cord = random.randint(1, 101)

      copy_for_bg_fg = copy.deepcopy(bg)
      copy_for_bg_flip_fg = copy.deepcopy(bg)
      copy_for_fg = copy.deepcopy(fg)
      copy_for_fg_mask = copy.deepcopy(fg_mask)

      # Creating black image to store the mask of bg_fg images
      black_imgage_for_fg_mask = Image.fromarray(black_image,mode='1')
      black_image_for_fg_flip_mask = Image.fromarray(black_image,mode='1')

      # Creating Flip images
      fg_flip_image = copy_for_fg.transpose(PIL.Image.FLIP_LEFT_RIGHT)
      fg_flip_mask_image = copy_for_fg_mask.transpose(PIL.Image.FLIP_LEFT_RIGHT)

      # Overlaying images
      copy_for_bg_fg.paste(copy_for_fg, (x_cord,y_cord),copy_for_fg)
      copy_for_bg_flip_fg.paste(fg_flip_image, (x_cord,y_cord),fg_flip_image)

      black_imgage_for_fg_mask.paste(copy_for_fg_mask, (x_cord,y_cord), copy_for_fg_mask)
      black_image_for_fg_flip_mask.paste(fg_flip_mask_image, (x_cord,y_cord), fg_flip_mask_image)

     
      copy_for_bg_fg.save("fg_bg_tmp.jpg",optimize=True, quality=30)
      zip_fg_bg.write('fg_bg_tmp.jpg', f'fg_bg{str(bg_fg_counter)}.jpg')

      bg_fg_counter += 1
     
      copy_for_bg_flip_fg.save("fg_bg_flip_tmp.jpg",optimize=True, quality=30)
      zip_fg_bg.write('fg_bg_flip_tmp.jpg', f'fg_bg{str(bg_fg_counter)}.jpg')


      
      black_imgage_for_fg_mask.save("fg_bg_mask_tmp.jpg",optimize=True, quality=30)
      zip_fg_bg_mask.write('fg_bg_mask_tmp.jpg', f"fg_bg_masks{str(bg_fg_mask_counter)}.jpg")

      bg_fg_mask_counter += 1
      
      black_image_for_fg_flip_mask.save('fg_bg_mask_flip_tmp.jpg',optimize=True, quality=30)
      zip_fg_bg_mask.write('fg_bg_mask_flip_tmp.jpg', f'fg_bg_masks{str(bg_fg_mask_counter)}.jpg')


      bg_fg_counter += 1
      bg_fg_mask_counter += 1

      
print(datetime.datetime.now())
```
